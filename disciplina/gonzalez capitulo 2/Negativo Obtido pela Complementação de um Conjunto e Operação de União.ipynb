##### Módulos utilizados

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

#### Imagem Original

```python
image = cv2.imread('imagens_originais/Fig0232(a)(partial_body_scan).tif')
plt.title('Imagem Original (a)')
plt.axis('off')
plt.imshow(image);
```

### Complemento da Imagem Original

Definição de conjunto: $A^c={(x,y,K-z)|(x,y,z)}$

Onde K é definido por: $2^k-1$,  sendo k o número de bits por pixel utilizado para representar a intensidade z.

```python
k = 8
K = 2**k -1
```

```python
# como o vetor é um array numpy que pode receber operações em todos o valores facilmente, 
# e os valores do vetor já são o próprio z da definição do conjunto, basta executar:
complement = K - image
```

```python
plt.title('Complemento da Imagem (b)')
plt.axis('off')
plt.imshow(complement);
```

### Arranjo Matricial Retangular

```python
imageB = image*0
```

```python
imageB = 3*(imageB+int(image.mean()))
```

```python
plt.title('Arranjo Matricial Retangular (r)')
plt.axis('off')
plt.imshow(imageB);
```

### Operação de União entre Imagem Original e Arranjo Matricial Retangular

```python
def union(imageA,imageB):
    rows, cols, _  = imageA.shape
    imageC = imageA*0
    for i in range(rows):
        for j in range(cols):
            if imageB[i][j][0] < imageA[i][j][0]:
                imageC[i][j] = imageA[i][j]
            else:
                imageC[i][j] = imageB[i][j]
    return imageC
```

```python
imageC = union(image,imageB)
```

```python
plt.title('Operação de União entre o Retangulo (r) e Imagem Original. (c)')
plt.axis('off')
plt.imshow(imageC);
```

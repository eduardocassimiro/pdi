### Operação de mudança do nível de cinza

Sabemos que a imagem possui 256 níveis de cinzam, ou seja, cada pixel pode variar entre $2^8$ valores.

Sabendo o qual o seu valor atual e a resolução máxima do nível seu nível de cinza, para realizarmos uma degressão no nível de cinza, basta realizarmos uma regra de três:

Definição de conjunto: $A^c={(x,y,K-z)|(x,y,z)}$

Onde K é definido por: $2^k-1$,  sendo k o número de bits por pixel utilizado para representar a intensidade z.

```python
teste = image.astype(np.uint16)
```

```python
image_k7 = (teste * (2**7))//(2**8)
```

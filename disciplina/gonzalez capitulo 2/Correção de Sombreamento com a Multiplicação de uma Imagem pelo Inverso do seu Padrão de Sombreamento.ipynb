##### Módulos utilizados

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

#### Imagem original

```python
image = cv2.imread('imagens_originais/Fig0229(a)(tungsten_filament_shaded).tif')

plt.axis("off")
plt.title("Imagem Original (a)")
plt.imshow(image);
```

#### Padrão de sombreamento

```python
shading = cv2.imread('imagens_originais/Fig0229(b)(tungsten_sensor_shading).tif')

plt.axis("off")
plt.title("Padrão de sombreamento (b)")
plt.imshow(shading);
```

### Produto da Imagem Original (a) pelo inverso do Padrão de Sombreamento (b)

```python
corrected = image * 1/shading
```

```python
plt.axis("off")
plt.title("Correção de sombreamento")
plt.imshow(corrected);
```

### Comparação

```python
fig = plt.figure(figsize=(10,5))

fig.add_subplot(1,2,1); plt.axis('off')
plt.title('Imagem Original')
plt.imshow(image)
fig.add_subplot(1,2,2); plt.axis('off')
plt.title(' Imagem com Sombreamento Corrigido')
plt.imshow(corrected)

plt.suptitle('Comparação das Imagens', fontsize=15)
plt.show();
```

##### Módulos utilizados

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from copy import deepcopy
```

```python
plt.style.use('ggplot')
```

## 1. Visualizando a imagem orignal

```python
# Abrindo a imagem
img = cv2.imread('imagens_originais/Fig3.12.tif')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Plotando a imagem original
plt.figure(figsize=(6,6))
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title('Imagem original: Angiograma da aorta')
plt.show()
```

## 2.  Fatiamento de níveis de intensidade
### 2.1 Seleção de intensidade

```python
# Definindo a amplitude máxima da imagem (Para facilitar a visualização dos cálculos)
L = 255
# Definindo 'uma parte' de acordo com a escala indicada do livro
L_part = L//5
# Valores de fatiamento de acordo com o livro
A = 2*L_part
B = 3*L_part + L_part//2
```

```python
# Aplicamendo na imagem
img3 = deepcopy(img)
for i in range(len(img)):
    for j in range(len(img[0])):
        if img[i][j] >= A-40 and img[i][j] <= B+20:
            img3[i][j] = 0
        else:
            img3[i][j] = 255
```

```python
# Plotando o resultado
plt.figure(figsize=(6,6))
plt.imshow(img3, cmap='gray')
plt.axis('off')
plt.title('Angiograma da aorta com a área selecionada ajustada para o preto')
plt.show()
```

### 2.2 Fatiamento de intensidade

```python
# Valores de fatiamento de acordo com o livro
A = L_part + L_part//2
B = 2*L_part + L_part//2
```

```python
# Aplicamendo na imagem
img2 = deepcopy(img)
for i in range(len(img)):
    for j in range(len(img[0])):
        if img[i][j] >= A and img[i][j] <= B+20:
            img2[i][j] = 0
```

```python
# Plotando o resultado
plt.figure(figsize=(6,6))
plt.imshow(img2, cmap='gray')
plt.axis('off')
plt.title('Angiograma da aorta com a área selecionada ajustada para o preto')
plt.show()
```

##### Módulos utilizados

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from copy import deepcopy
from math import log
```

```python
plt.style.use('ggplot')
```

## 1. Visualizando a imagem e o seu histograma

```python
# Abrindo a imagem
img = cv2.imread('imagens_originais/Fig3.5.jpg')
# Transformando a imagem colorida em cinza
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title('Espectro de Fourier')
plt.show()
```

## 2. Transformações logarítmicas

A forma geral da transformação logarítmica é <br>
<center>$s = c log (1 + r)$</center>
onde $c$ é uma constante e $r$ é o valor de cinza do referido pixel (considera-se que r ≥ 0).

```python
# Copiando a imagem
img_log = deepcopy(img)
```

```python
# Utilizaremos a constante c = 1 como indicado pelo livro para esse exemplo
c = 1
```

```python
# Realizando o cálculo pixel a pixel
for i in range(len(img_log)):
    for j in range(len(img_log[0])):
        img_log[i][j] = c*log(1+img_log[i][j])
```

```python
plt.imshow(img_log, cmap='gray')
plt.axis('off')
plt.title('Resultado da aplicação da Transformação Logarítmica no Espectro de Fourier')
plt.show()
```

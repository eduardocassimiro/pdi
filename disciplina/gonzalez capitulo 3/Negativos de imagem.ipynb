##### Módulos utilizados

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
from copy import deepcopy
```

```python
plt.style.use('ggplot')
```

## 1. Visualizando a imagem e o seu histograma

```python
# Abrindo a imagem
img = cv2.imread('imagens_originais/Fig3.4.jpg')
# Transformando a imagem colorida em cinza
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Calculando o histograma da imagem
hist = cv2.calcHist(img,[0],None,[256],[0,256])
```

```python
# Plotando a a imagem e o seu histograma
fig = plt.figure(figsize=(14,7))

fig.add_subplot(1,2,1); plt.axis('off')
plt.title('Imagem Original')
plt.imshow(img, cmap='gray')

fig.add_subplot(1,2,2);
plt.title('Histograma')
plt.plot(hist,color='black',lw=1)

fig.suptitle('Mamografia Digital', fontsize=20)
plt.show()
```

## 2.  Negativo da imagem obtido pela inversão da itensidade

O negativo de uma imagem com níveis de intensidade na faixa [0, L – 1] é obtido utilizando a transformação de negativo mostrada na Figura 3.3, dada pela expressão: <br>
<center>$s = L -1 - r$</center> 
onde $L$ é a amplitude máxima da cor (256), o $-1$ para ajustar o a escala das cores entre 0 e 255 e $r$ é o valor de cinza do pixel.

```python
# Copiando a imagem
img_inv = deepcopy(img)
```

```python
# Calculando a amplitude máxima do RGB
L = 255 + 1
```

```python
# Realizando o cálculo pixel a pixel
for i in range(len(img_inv)):
    for j in range(len(img_inv[0])):
        img_inv[i][j] = L - 1 - img_inv[i][j]
```

```python
# Calculando o histograma da imagem invertida
hist_inv = cv2.calcHist(img_inv,[0],None,[256],[0,256])
```

```python
fig = plt.figure(figsize=(14,7))

fig.add_subplot(1,2,1); plt.axis('off')
plt.title('Negativo da Imagem Original')
plt.imshow(img_inv, cmap='gray')

fig.add_subplot(1,2,2);
plt.title('Histograma')
plt.plot(hist_inv,color='black',lw=1)

fig.suptitle('Negativo da Mamografia Digital', fontsize=20)
plt.show()
```
